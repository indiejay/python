```python
def input1(A,x,y):
    if A[x][y] == 0:
        A[x][y] = "X"
    return A 

def input2(A,x,y):
    if A[x][y] == 0:
        A[x][y] = "O"
    return A

def grid_check(A):
    for i in range(len(A)):
        if A[i][0] == A[i][1]==A[i][2] and A[i][0] != 0:
            return A[i][0]
    for i in range(len(A)):
        if A[0][i]==A[1][i]==A[2][i] and A[0][i] != 0:
            return A[0][i]    
    if A[0][0]==A[1][1]==A[2][2] and A[0][0] != 0:
        return A[0][0]                   
    elif A[0][2] == A[1][1] == A[2][0] and A[0][2]!=0:
        return A[0][2]
    
def print_grid(A):
    for i in A:
        print(i)



A = [[0,0,0],[0,0,0],[0,0,0]]
x = 0
y = 0
i = 0
while i != 3:
    while True:
        element = (input("player1 enter the row and column seperated by comma: ").split(","))
        x = int(element[0])-1
        y = int(element[1])-1
        if A[x][y] != 0:
            print("this is already take Enter again: ")
            continue
        else:
            input1(A,x,y)
            break
    while True:
        element2 = (input("player2 enter the row and column seperated by comma: ").split(","))
        x = int(element2[0])-1
        y = int(element2[1])-1
        if A[x][y] !=0:
            print("this is already take Enter again: ")
            continue
        else:
            input2(A,x,y)
            break
    i+=1
print_grid(A)
winner=grid_check(A)
if winner == "X":
    print("Palyer 1 wins")
elif winner == "O":
    print("Player 2 wins")
else:
    print("no one wins")
```
